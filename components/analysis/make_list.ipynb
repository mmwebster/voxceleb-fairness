```python
import pandas as pd
```

```python
meta = pd.read_csv('vox1_meta.csv', sep='\t')
```

```python
meta
```

```python
def GetIds(male=True):
    g = 'm' if male else 'f'
    return meta[meta["Gender"] == g]['VoxCeleb1 ID']
```

```python
male_ids = set(GetIds())
female_ids = set(GetIds(male=False))
```
